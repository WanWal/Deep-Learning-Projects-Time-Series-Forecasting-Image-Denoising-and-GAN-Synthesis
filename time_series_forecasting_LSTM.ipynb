# Cleaning data & preprocessing

```python
# Hitung persentase missing per kolom
df.isnull().sum() / len(df) * 100
```

Setelah melihat persentasi nilai nan yang ada, maka untuk tiap kolom yang memiliki > 40 - 50% nilai NAN, akan dilakukan mendrop kolom tersebut. Dan untuk yang dibawah, nilai NAN akan di fill dengan value lainnya

```python
print(df.dtypes[['From Date','To Date']])
```

Mengecheck Dtype dari kolom date

```python
# Mengubah kolom tanggalnya ke datetime agar model paham
df['From Date'] = pd.to_datetime(df['From Date'])
df['To Date']   = pd.to_datetime(df['To Date'])

# Melakukan set index dengan kolom 'From Date'
df = df.set_index('From Date')
df = df.drop(columns='To Date')
df = df.asfreq('H')

# Drop kolom yang >50% datanya hilang
batas = 0.5 * len(df)
to_drop = [col for col in df.columns if df[col].isna().sum() > batas]
df = df.drop(columns=to_drop)

# Update
numeric = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Isi nilai NAN menggunakan Interpolasi time-based pada kolom numerik
df[numeric] = df[numeric].interpolate(method='time', limit=3)

# Isi nilai NAN dengan forward/backward fill
df[numeric] = df[numeric].fillna(method='ffill').fillna(method='bfill')

df.isna().sum()
```

Codingan diatas dilakukan untuk mencapai hal ini:  
- Mengubah format “From Date” dan “To Date” sebagai nilai waktu, bertujuan untuk operasi deret waktu (interpolasi, reindexing, dsb.) sehingga model akan dapat berjalan dengan benar.
- Melakukan drop kolom "To Date", karena dengan adanya index menggunakan "From Date" kita tidak perlu lagi to date karena waktu sudah tercatat di kolom "From Date".
- Dengan melakukan set_index akan membuat cap waktu (timestamp) menjadi label baris utama.
- asfreq akan menambahkan baris kosong untuk nilai (NaN), yang bertujuan inputan dari nilai NAN akan sesuai dengan interval baris tersebut dimana intervalnya yaitu tepat 1 jam.  

Alasan:
Dikarenakan model deret waktu (sliding-window, LSTM, ARIMA, dll.) mensyaratkan data yang digunakan di interval yang konsisten, tanpa adanya “lompatan” waktu. Jadi perlu di sebutkan rentang batasnya.

```python
# Chek adakah outliers
for col in numeric:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Clip nilai ke dalam [lower_bound, upper_bound]
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)
```

Karena banyaknya outliers yang ada pada tiap kolom, perlu dilakukannya penanganan outliers-outliers tersebut sehingga model tidak akan melenceng prediksinya.

```python
# Split Data menjadi (80/10/10):
n = len(df)
train_df = df.iloc[: int(0.8 * n)]
val_df   = df.iloc[int(0.8 * n) : int(0.9 * n)]
test_df  = df.iloc[int(0.9 * n) : ]
```

Disini dilakukannya split data dengan menggunakan iloc, dengan rentang 80%, 10%, 10%. Dikarenakan tidak mengetahui rentang waktu pada data secara jelas. iloc cocok digunakan tanpa memikirkan index yang akan kita pilih.

```python
len(train_df)
```

```python
from sklearn.preprocessing import MinMaxScaler
# Scaling dengan MinMaxScaler (fit di train, transform di val/test):
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_df)
val_scaled   = scaler.transform(val_df)
test_scaled  = scaler.transform(test_df)
```

Pada train digunakannya fit_transform, bertujuan untuk menyesuaikan rentang data (min/max) nya. dengan scale [0,1].
Untuk val dan test tidak digunakannya fit agar menghindari leaked data, meskipun masih sama-sama di scale [0,1].

Normalisasi fitur penting dilakukan untuk model ML/DL agar setiap variabel berada di skala serupa sehingga dapat mempercepat konvergensi dan mencegah adanya rentang data yang besar.

```python
def windows(data, input_steps=5, output_steps=1, target_col='AT (degree C)'):
    X, y = [], []
    arr = np.array(data)
    tgt_idx = data.columns.get_loc(target_col)
    for i in range(len(arr) - input_steps - output_steps + 1):
        X.append(arr[i : i + input_steps])
        y.append(arr[i + input_steps : i + input_steps + output_steps, tgt_idx])
    return np.array(X), np.array(y)
```

```python
X_train, y_train = windows(pd.DataFrame(train_scaled, columns=train_df.columns))
X_val,   y_val   = windows(pd.DataFrame(val_scaled,   columns=val_df.columns))
X_test,  y_test  = windows(pd.DataFrame(test_scaled,  columns=test_df.columns))

y_train = y_train.squeeze(); y_val = y_val.squeeze(); y_test = y_test.squeeze()
```

Di tahap ini, dibuatkan function windows, bertujuan agar model melakukan input 5 baris berututan (5 jam sebelumnya) sehingga model akan mampu memprediksi kondisi di 1 jam kedepan dan memastikan tidak adanya lompatan waktu yang akan digunakan di selanjutnya.  


Dilakukan squeeze agar memastikan bentuk array label (1D) agar sesuai dengan ekspektasi model dan fungsi evaluasi sehingga dapat mencegah bug dan membuat pipeline training/evaluasi lebih bersih.

# Modeling

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Definisi model
class BaselineLSTM(nn.Module):
    def __init__(self, input_dim, hidden_size=10):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden_size,
                            num_layers=1,
                            batch_first=True)
        self.regressor = nn.Linear(hidden_size, 1)

    def forward(self, x):
        _, (hn, _) = self.lstm(x)
        return self.regressor(hn[-1])

# Setup device, model, loss, optimizer
device    = torch.device("cuda" if torch.cuda.is_available() else "cpu")
input_dim = X_train.shape[2]
model     = BaselineLSTM(input_dim).to(device)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

# Buat DataLoader
def get_loader(X, y, batch_size=32, shuffle=True):
    ds = TensorDataset(
        torch.tensor(X, dtype=torch.float32),
        torch.tensor(y, dtype=torch.float32).unsqueeze(1)
    )
    return DataLoader(ds, batch_size=batch_size, shuffle=shuffle)
```

```python
# Inisialisasi model, loss, optimizer, dan dataloader
device    = torch.device("cuda" if torch.cuda.is_available() else "cpu")
input_dim = X_train.shape[2]
model     = BaselineLSTM(input_dim).to(device)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)


train_loader = get_loader(X_train, y_train, batch_size=32, shuffle=True)
val_loader   = get_loader(X_val,   y_val,   batch_size=32, shuffle=False)
test_loader  = get_loader(X_test,  y_test,  batch_size=32, shuffle=False)

# Training
n_epochs = 20
model.train()
for epoch in range(1, n_epochs + 1):
    running_loss = 0.0
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        preds = model(xb)
        loss  = criterion(preds, yb)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * xb.size(0)
    epoch_loss = running_loss / len(train_loader.dataset)
    print(f"Epoch {epoch:02d} | Train Loss: {epoch_loss:.4f}")

# Simpan model
torch.save(model.state_dict(), "baseline_model.pt")
```

Model LSTM kali ini menunjukkan adanya penurunan train loss yang signifikan dari 0.0176 di epoch pertama menjadi 0.0022 di epoch kedua, lalu menjadi stabil di kisaran 0.0012 hingga 0.0011 dimulai pada epoch ke-5 hingga ke-20.   
Hal ini menunjukkan bahwa model berhasil/mampu belajar dengan baik dan telah mencapai konvergensi.

```python
# Load model untuk test
model = BaselineLSTM(input_dim).to(device)
model.load_state_dict(torch.load("baseline_model.pt"))
model.eval()

# Buat fungsi evaluasi untuk test model
def evaluate_model(model, dataloader, name="Model"):
    model.eval()
    all_preds, all_targets = [], []

    with torch.no_grad():
        for xb, yb in dataloader:
            xb = xb.to(device)
            preds = model(xb).cpu().numpy()
            all_preds.append(preds)
            all_targets.append(yb.numpy())

    y_true = np.concatenate(all_targets).flatten()
    y_pred = np.concatenate(all_preds).flatten()

    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    r2  = r2_score(y_true, y_pred)

    print(f"\n{name}")
    print(f"MAE : {mae:.4f}")
    print(f"MSE : {mse:.4f}")
    print(f"R²  : {r2:.4f}")

    return mae, mse, r2
```

```python
# Evaluasi model di test set
evaluate_model(model, test_loader, "Baseline LSTM (Test Data)")
```

# Modifikasi model lstm

```python
# Model LSTM dengan Dropout
class LSTMWithDropout(nn.Module):
    def __init__(self, input_dim, hidden_size=10, dropout=0.3):
        super().__init__()
        self.lstm = nn.LSTM(input_dim,
                            hidden_size,
                            num_layers=1,
                            batch_first=True)
        self.dropout = nn.Dropout(dropout)  # dropout
        self.regressor = nn.Linear(hidden_size, 1)

    def forward(self, x):
        _, (hn, _) = self.lstm(x)
        h = self.dropout(hn[-1])
        return self.regressor(h)
# Setup device, model, loss, optimizer
device    = torch.device("cuda" if torch.cuda.is_available() else "cpu")
input_dim = X_train.shape[2]
model     = LSTMWithDropout(input_dim=input_dim,
                            hidden_size=10,
                            dropout=0.3).to(device)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)

# Fungsi DataLoader
def get_loader(X, y, batch_size=32, shuffle=True):
    ds = TensorDataset(torch.tensor(X, dtype=torch.float32),
                       torch.tensor(y, dtype=torch.float32).unsqueeze(1))
    return DataLoader(ds, batch_size=batch_size, shuffle=shuffle)
```

```python
train_loader = get_loader(X_train, y_train, batch_size=32, shuffle=True)
val_loader   = get_loader(X_val,   y_val,   batch_size=32, shuffle=False)
test_loader  = get_loader(X_test,  y_test,  batch_size=32, shuffle=False)

# Training
n_epochs = 20
model.train()
for epoch in range(1, n_epochs+1):
    running_loss = 0.0
    for xb, yb in train_loader:
        xb, yb = xb.to(device), yb.to(device)

        optimizer.zero_grad()
        preds = model(xb)
        loss  = criterion(preds, yb)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * xb.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    print(f"Epoch {epoch:02d} | Train Loss: {epoch_loss:.4f}")

# Simpan model
torch.save(model.state_dict(), "modifikasi_model.pt")
```

Model mampu menunjukkan adanya penurunan train loss dari 0.0257 ke 0.0028 hanya dalam 20 epoch, menandakan adanya penurunan tajam di awal dan stabil setelah epoch ke-10, menunjukkan bahwa model berhasil/mampu belajar dengan baik dan mencapai konvergensi.  
Diberikan dropout juga membantu model mencegah adanya overfitting meski membuat loss sedikit lebih tinggi dari baseline.

```python
# Evaluasi Model di Test Set
def evaluate_model(model, dataloader, name="Model"):
    model.eval()
    all_preds, all_targets = [], []

    with torch.no_grad():
        for xb, yb in dataloader:
            xb = xb.to(device)
            preds = model(xb).cpu().numpy()
            all_preds.append(preds)
            all_targets.append(yb.numpy())

    y_true = np.concatenate(all_targets).flatten()
    y_pred = np.concatenate(all_preds).flatten()

    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    r2  = r2_score(y_true, y_pred)

    print(f"\n== {name} ==")
    print(f"MAE : {mae:.4f}")
    print(f"MSE : {mse:.4f}")
    print(f"R²  : {r2:.4f}")

    return mae, mse, r2

# Load ulang dan evaluasi
modif_model = LSTMWithDropout(input_dim=input_dim, hidden_size=10, dropout=0.3).to(device)
modif_model.load_state_dict(torch.load("modifikasi_model.pt"))

evaluate_model(modif_model, test_loader, "Modified LSTM with Dropout (Test Data)")
```

Modifkasi model LSTM yang diberikan dropout berhasil meningkatkan akurasi prediksi. Penurunan MAE dan MSE menunjukkan error yang lebih kecil, sementara kenaikan R² menandakan model menjelaskan variansi data lebih baik. Dropout terbukti efektif mencegah overfitting dan meningkatkan generalisasi.

## Hyperparameter Tune

```python

# Definisi ulang model untuk fleksibilitas
class HypertuneModel(nn.Module):
    def __init__(self, input_dim, hidden_size=10, dropout=0.3):
        super().__init__()
        self.lstm = nn.LSTM(input_dim, hidden_size, num_layers=1, batch_first=True)
        self.dropout = nn.Dropout(dropout)
        self.regressor = nn.Linear(hidden_size, 1)

    def forward(self, x):
        _, (hn, _) = self.lstm(x)
        h = self.dropout(hn[-1])
        return self.regressor(h)

# Fungsi DataLoader
def get_loader(X, y, batch_size=32, shuffle=True):
    ds = TensorDataset(torch.tensor(X, dtype=torch.float32),
                       torch.tensor(y, dtype=torch.float32).unsqueeze(1))
    return DataLoader(ds, batch_size=batch_size, shuffle=shuffle)

train_loader = get_loader(X_train, y_train)
val_loader   = get_loader(X_val, y_val)

# Hyperparameter grid
hidden_sizes = [10, 20]
dropouts = [0.2, 0.3]
learning_rates = [1e-3, 1e-4]
n_epochs = 10
input_dim = X_train.shape[2]
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
best_val_loss = float("inf")

best_model = None
best_params = {}

# Grid Search Loop
for hs in hidden_sizes:
    for do in dropouts:
        for lr in learning_rates:
            print(f"Training: hidden_size={hs}, dropout={do}, lr={lr}")
            model = LSTMWithDropout(input_dim=input_dim, hidden_size=hs, dropout=do).to(device)
            optimizer = torch.optim.Adam(model.parameters(), lr=lr)
            criterion = nn.MSELoss()

            # Training loop
            model.train()
            for epoch in range(n_epochs):
                for xb, yb in train_loader:
                    xb, yb = xb.to(device), yb.to(device)
                    optimizer.zero_grad()
                    preds = model(xb)
                    loss = criterion(preds, yb)
                    loss.backward()
                    optimizer.step()

            # Validation loss
            model.eval()
            val_loss = 0.0
            with torch.no_grad():
                for xb, yb in val_loader:
                    xb, yb = xb.to(device), yb.to(device)
                    preds = model(xb)
                    loss = criterion(preds, yb)
                    val_loss += loss.item() * xb.size(0)
            val_loss /= len(val_loader.dataset)

            print(f"→ Val Loss: {val_loss:.4f}")

            if val_loss < best_val_loss:
                best_val_loss = val_loss
                best_model = model
                best_params = {'hidden_size': hs, 'dropout': do, 'lr': lr}

# Simpan model terbaik
torch.save(best_model.state_dict(), "best_lstm_dropout_model.pt")
print(f"\nBest Params: {best_params}, Val Loss: {best_val_loss:.4f}")
```

Dari hasil diatas, model belajar dengan mencoba hidden_size, learing rate dan dropout yang berbeda sehingga menghasilkan loss yang yang beragam. Namun hasil terbaik pada hidden_size = 20, drop out =  03 dan learning rate pada 0.001 dengan hasil val loss terkecil yaitu 0.0011.

## Evaluasi secara dekat

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate_model(model, dataloader, name="Model"):
    model.eval()
    all_preds = []
    all_targets = []

    with torch.no_grad():
        for xb, yb in dataloader:
            xb, yb = xb.to(device), yb.to(device)
            preds = model(xb)
            all_preds.append(preds.cpu())
            all_targets.append(yb.cpu())

    y_true = torch.cat(all_targets).numpy()
    y_pred = torch.cat(all_preds).numpy()

    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    r2  = r2_score(y_true, y_pred)

    print(f"== {name} ==")
    print(f"MAE : {mae:.4f}")
    print(f"MSE : {mse:.4f}")
    print(f"R²  : {r2:.4f}")
    print("-" * 30)
    return mae, mse, r2
```

# Pertanyaan Bonus,   
apakah problem ini mungkin menggunakan time series multivariate, bagaimana mengetahui apakah problem tersebut dapat menjadi multivariate, jelaskan jawaban anda untuk problem ini

Masalah ini bisa menjadi time series multivariate, jika kita menggunakan dataset yang memiliki lebih dari satu fitur input yang dapat berubah dari waktu ke waktu.

Contohnya:   
jika kita melakukan prediksi harga emas, variabel yang kita gunakan selain adanya data harga sebelumnya. Kita juga perlu mempunyaii data seperti nilai tukar, inflasi, atau indeks saham, dengan beberapa variabel itulah disebut multivariate, karena ada banyak variabel yang ikut memengaruhi prediksi.

Cara mengeceknya:

Cek bentuk X_train:   
jika bentuknya (jumlah_data, panjang_waktu, jumlah_fitur) dan jumlah_fitur lebih dari 1, berarti data kita sudah multivariate.

Kalau hanya 1 fitur (univariate), Kita juga masih bisa mengubahnya menjadi multivariate dengan melakukan penambahan fitur lain yang relevan dan berubah terhadap waktu.
